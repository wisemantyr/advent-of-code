```python
def count_letter (letters, members):
    
    checked_letters = []
    yes = 0
    
    for letter in letters:
        if letter in checked_letters:
            continue
        else:
            checked_letters.append(letter)
            count = letters.count(letter)
            if count == members:
                yes += 1
    print(f"number questions unanimous: {yes}")
    print("-" * 50)
    unanimous.append(yes)
    return unanimous
```
