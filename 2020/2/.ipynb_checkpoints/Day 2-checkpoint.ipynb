```python
count_valid = 0
for index in input_list:
    split1 = re.split("-", index)
    split2 = re.split("\s", split1[1])
    split2[1] = split2[1].strip(":")
    
    pos1 = int(split1[0]) - 1
    pos2 = int(split2[0]) - 1
    letter = split2[1]
    password = split2[2]

    
    if password[pos1] == letter and password[pos2] != letter:
        print("Valid")
        count_valid = count_valid + 1
    elif password[pos1] != letter and password[pos2] == letter:
        print("Valid")
        count_valid = count_valid + 1
```
