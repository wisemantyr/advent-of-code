```python
def seat_row(seats):
    max_row = 127
    min_row = 0
    row_range = (min_row, max_row)
    
    for seat in seats[0:7]:
        for char in seat:
            print(char)
            if char == 'F':
                print(f"start row range: {row_range}")
                
                max_row = row_range[1]
                min_row = row_range[0]
                max_row = math.floor(((max_row-min_row)/2) + min_row)
                print(f"new max row: {max_row}")
                
                row_range = (min_row, max_row)
                print(f"new row_range: {row_range}")
            else:
                print(f"start row range: {row_range}")
                
                max_row = row_range[1]
                min_row = row_range[0]
                min_row = round(((max_row - min_row)/2) + min_row)
                print(f"new min row: {min_row}")
                
                row_range = (min_row, max_row)
                print(f"new row range: {row_range}")
                
    return max_row
```

```python
row = seat_row(inp[0])
```

```python
def seat_col(seats):
    max_col = 7
    min_col = 0
    col_range = (min_col, max_col)
    
    for seat in seats[7:]:
        for char in seat:
            print(char)
            if char == 'L':
                print(f"start col range: {col_range}")
                
                max_col = col_range[1]
                min_col = col_range[0]
                max_col = math.floor(((max_col-min_col)/2) + min_col)
                print(f"new max col: {max_col}")
                
                col_range = (min_col, max_col)
                print(f"new col_range: {col_range}")
            else:
                print(f"start col range: {col_range}")
                
                max_col = col_range[1]
                min_col = col_range[0]
                min_col = round(((max_col - min_col)/2) + min_col)
                print(f"new min col: {min_col}")
                
                col_range = (min_col, max_col)
                print(f"new col range: {col_range}")
                
    return max_col
```

```python
col = seat_col(inp[0])
```

```python
def get_id(row,col):
    return (row * 8) + col

get_id(row, col)
```

```python
seats = []

for seat in inp:
    
    row = seat_row(seat)
    col = seat_col(seat)
    
    seat_id = get_id(row,col)
    
    seat_info = {}
    seat_info['id'] = seat_id
    seat_info.setdefault('seat', (row, col))
    
    seats.append(seat_info)

seats
```

### Part 2

It's a completely full flight, so your seat should be the only missing boarding pass in your list. However, there's a catch: some of the seats at the very front and back of the plane don't exist on this aircraft, so they'll be missing from your list as well.

Your seat wasn't at the very front or back, though; the seats with IDs +1 and -1 from yours will be in your list.

```python
total_assigned = len(seats) + 1
potential_plane = (last_row + 1) * (last_col + 1)
first_last = (last_col + 1) * 2
potential_mine = (potential_plane - total_assigned) - first_last

print(f"total possible seats on plane: {potential_plane}")
print(f"total assigned seats on plane: {total_assigned}")
print(f"num seats potentially mine: {potential_mine}")
```

```python
assigned = [seat['seat'] for seat in seats]
rows_assigned =[x for (x,y) in assigned]
```

```python
high_row_seated = max(rows_assigned)
low_row_seated = min(rows_assigned)

print(f"lowest row assigned: {low_row_seated}")
print(f"highest row assigned: {high_row_seated}")
```

```python
unassigned_front = (low_row_seated) * (last_row + 1)
unassigned_back = ((last_row + 1) - high_row_seated) * (last_row + 1)

unassigned = unassigned_front + unassigned_back
possible_unassigned = potential_plane - total_assigned

print(f"number unassigned possible: {possible_unassigned}")
```
