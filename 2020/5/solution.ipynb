```python
def front(seat_range):
    print(f"start row range: {seat_range}")
    
    first = seat_range[0]
    last = math.floor(((seat_range[1]-first)/2) + first)
    print(f"new max row: {last}")
    
    return (first, last)
    print(f"({first}, {last})")

def back(seat_range):
    print(f"start row range: {seat_range}")
    
    last = seat_range[1]
    first = round(((last - seat_range[0])/2) + seat_range[0])
    print(f"new min row: {first}")
    
    return (first, last)
    print(f"({first}, {last})")


def seat_row(seats):
    last = 127
    first = 0
    seat_range = (first, last)
    
    for seat in seats[0:7]:
        for char in seat:
            print(char)
            if char == 'F':
                seat_range = front(seat_range)
            else:
                seat_range = back(seat_range)
    return seat_range[1]
```

```python
def get_row(first, last, seats):
    seat_range = (first, last)
    for letters in seats[0:7]:
        for index in letters:
            if index == 'F':
                seat_range = front(seat_range)
            else:
                seat_range = back(seat_range)
    row = seat_range[1]
    return row
row = get_row(first_row, last_row, inp[0])
row
```

```python
def seat_col(seats):
    max_col = 7
    min_col = 0
    col_range = (min_col, max_col)
    
    for seat in seats[7:]:
        for char in seat:
            print(char)
            if char == 'L':
                print(f"start col range: {col_range}")
                
                max_col = col_range[1]
                min_col = col_range[0]
                max_col = math.floor(((max_col-min_col)/2) + min_col)
                print(f"new max col: {max_col}")
                
                col_range = (min_col, max_col)
                print(f"new col_range: {col_range}")
            else:
                print(f"start col range: {col_range}")
                
                max_col = col_range[1]
                min_col = col_range[0]
                min_col = round(((max_col - min_col)/2) + min_col)
                print(f"new min col: {min_col}")
                
                col_range = (min_col, max_col)
                print(f"new col range: {col_range}")
                
    return max_col
```

```python
def check_surrounding(missing_seats, seated):
    for a in missing_seats:
        count_true = 0
        
        seat_up = (a[0] + 1, a[1])
        seat_down = (a[0] - 1, a[1])
        seat_left = (a[0], a[1] + 1)
        seat_right = (a[0], a[1] - 1)
        
        check_seats = [seat_up, seat_down, seat_left, seat_right]
        
        for b in check_seats:
            for c in seated:
                if b == c:
                    count_true += 1
                    print(f"{b}: True")
        if count_true == 4:
            return a

check_surrounding(missing_seats, seated)
        
```
