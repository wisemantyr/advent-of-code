```python
def parse_tups(target, li):
    is_target = False
    for (x,y) in li:
        if x == target:
            is_target = True
    return is_target
```

```python
bags = []
count= 0

def get_gold(d, target):
    global count
    global bags
    targets = {'target': target}
    print(f"checking: {targets['target']}")
    keys = list(d)
    next_targets = []
    for k in keys:
        items = list(d[k].items())
        if parse_tups(target, items):
            next_targets.append(k)
            bags.append(k)
            count += 1
            print(f"{k} CONTAINS {target} !")
    bags = list(set(bags))
    count = len(bags)
    print("")
    print(f"count: {count}")
    print("")
    print(f'next targets: {next_targets}')
    if next_targets:
        print("")
        print("---" * 40)
        print("---" * 40)
        print("")
        for t in next_targets:
            print("")
            get_gold(d, t)
            print("")
            print(("---" * 5)+ "stopped checking"+ " "+ t + " " + ("---" * 5))
            print("")
    return count
    
num_bags = get_gold(bag_rules, 'shiny gold')
num_bags
```
