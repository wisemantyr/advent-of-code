```python
with open('input.txt', 'r') as f:
    input_list = [line.strip() for line in f]
input_list
```

```python
pattern_len = len(input_list[0])
pattern_len
```

```python
input_len=len(input_list)
input_len
```

```python
room = len(input_list[0]) - 1
room
```

```python
for (right, down) in slopes:
    moves = math.floor(room/right)
    iters = math.floor((input_len - down)/moves)
    repeated_input = [(line*iters) for line in input_list]
    print(f"Slope: ({right}, {down})")
    print(f"Number of moves before string ends: {moves}")
    print(f"Number of rows before string ends: {iters}")
```

```python
trees = []
for (right, down) in slopes:
    tree_count = 0
    moves = math.floor(room/right)
    iters = round((input_len - down)/moves)
    repeated_input = [(line*iters) for line in input_list]
    row_num = down
    index = right
    forest = repeated_input[down::down]
    print(f"Slope: ({right}, {down})")
    print(f"Number of moves before string ends: {moves}")
    print(f"Number of rows before string ends: {iters}")
    print(f"Length of row: {len(forest[index])}")
    print(f"Number of rows: {len(forest)}")
    for row in forest:
        print(f"{row_num}: {row[index]}")
        if row[index] == "#":
            tree_count = tree_count + 1
        index += right
        row_num += down
    trees.append(tree_count)
    print(f"Number of trees: {tree_count}")
    print("---"*20)
```
